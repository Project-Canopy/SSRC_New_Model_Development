```python
# reset index first

tags_dict = {}

for i in range(len(osm_data)):
    tags = osm_data.loc[i, 'tags']
    for k in tags.keys():
        v = tags[k]
        if type(v) is str:
            tag_str = f'{k}_{v}'
            if tag_str in tags_dict.keys():
                tags_dict[tag_str] += 1
            else:
                tags_dict[tag_str] = 1
        elif type(v) is list:
            for val in v:
                tag_str = f'{k}_{val}'
                if tag_str in tags_dict.keys():
                    tags_dict[tag_str] += 1
                else:
                    tags_dict[tag_str] = 1
        else:
            print('Unexpected value:', v)
            continue
            
print('Finished')
```
