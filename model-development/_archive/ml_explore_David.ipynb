```python
def make_labels_df(filepath):

    labels_df = pd.read_csv(filepath)

    label_list = ['Habitation', 'ISL', 'Industrial_agriculture', 'Mining',
        'Rainforest', 'River', 'Roads', 'Savannah', 'Shifting_cultivation',
        'Water'
    ]
    
    labels_df.columns = label_list + ['path']
    
    return labels_df
```

```python
def extract_raster(tiff_filename, bands=['all']):
    with rio.open(tiff_filename) as src:
        if bands == ['all']:
            return src.read()
        else:
            return src.read(bands)
```

```python
def test_generator(input_a):
    
    while True:
        ragged_tensor = tf.ragged.constant([[1, 2], [3, 4]])
    
        yield input_a, ragged_tensor
```

```python
dataset = tf.data.Dataset.from_generator(
     test_generator, args=([1],[2]),
     output_types=tf.int32, output_shapes = (),)
```

```python
def generate_dataset():
    
    s3 = boto3.resource('s3')
    bucket_name = 'canopy-production-ml'
    labels_df = make_labels_df('D:/canopy_data/csvs/labels_test_v1.csv')
    
    for i in labels_df.index:
        file_path = labels_df.loc[i, 'paths']
        obj = s3.Object(bucket_name, file_path)
        obj_bytes = io.BytesIO(obj.get()['Body'].read())
        raster = extract_raster(obj_bytes)
        
        label_list = ['Habitation', 'ISL', 'Industrial_agriculture', 'Mining',
            'Rainforest', 'River', 'Roads', 'Savannah', 'Shifting_cultivation',
            'Water'
        ]
        
        y = []
        for label in label_list:
            y.append(labels_df.loc[i, label])
            
        yield raster, np.asarray(y)
```

```python
labels_df = make_labels_df('D:/canopy_data/csvs/labels_test_v1.csv')

labels_df.head()
```

```python
dataset = tf.data.Dataset.from_generator(
     generate_dataset,
     output_types=tf.int16)
```
