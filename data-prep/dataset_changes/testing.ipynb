```python
error_paths = []

for i in range(len(df1)):
    print(f'Row {i+1} of {len(df1)}', end='\r', flush=True)
    df_row = df1.iloc[[i]]
    merged = df_row.merge(df3, on=['0', '1', '2', '3', '4', 'paths'])
    if len(merged) == 0:
        path = df1.loc[0, 'paths']
        print('Error on path', path)
        error_paths.append(path)
```

```python
df_train = pd.read_csv('new_train_labels_v1.csv')
df_val = pd.read_csv('new_val_labels_v1.csv')
df_test = pd.read_csv('new_test_labels_v1.csv')
```

```python
df = pd.concat([df_train, df_val])
df = pd.concat([df, df_test])

assert len(df) == len(df_train) + len(df_val) + len(df_test)
```
